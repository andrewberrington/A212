```python
# Pine Chapter 5, Problem 1

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-1,3,40)
y = np.square(x)*3

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function y = 3x^2 for -1 <= x <= 3')
plt.show()
```

```python
# Pine Chapter 5, Exercise 2

x2 = np.linspace(-15,15,500)
y2 = np.cos(x2)/(1+((1/5)*np.square(x2)))

plt.plot(x2,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function y = cos(x)/(1 + 1/5(x^2)) for -15 <= x <= 15')
plt.show()
```
