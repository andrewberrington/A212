```python
import PIL
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import numpy as np
from pathlib import Path
import a212data
%matplotlib inline
datadir = a212data.__path__[0]
stinkpath = Path(datadir).joinpath('stinkbug.png')
img=mpimg.imread(str(stinkpath))
imgplot = plt.imshow(img)
lum_img = img[:,:,0]
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('hot')
plt.grid(False)
plt.show()
```

```python
# New plots

plt.clf()
cmap1 = sns.light_palette("cornflowerblue",as_cmap=True)
plt.cm.register_cmap(name='cfbl', cmap=cmap1)
imgplot2 = plt.imshow(lum_img)
imgplot2.set_cmap('cfbl')
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
plt.clf()
cmap2 = sns.dark_palette("red",as_cmap=True)
plt.cm.register_cmap(name='rd', cmap=cmap2)
imgplot3 = plt.imshow(lum_img)
imgplot3.set_cmap('rd')
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
plt.clf()
cmap3 = sns.cubehelix_palette(6, start=0.25, rot=-.9,as_cmap=True)
plt.cm.register_cmap(name='ch1', cmap=cmap3)
imgplot4 = plt.imshow(lum_img)
imgplot4.set_cmap('ch1')
plt.colorbar()
plt.grid(False)
plt.show()
```
