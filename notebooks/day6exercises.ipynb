```python
import numpy as np
import math as ma
from matplotlib import pyplot as plt
%matplotlib inline

# Exercise 3

xvals = np.linspace(-np.pi,np.pi,200)
singraph = np.sin(xvals)
cosgraph = np.cos(xvals)

fig,ax = plt.subplots(1,1)
ax.plot(xvals,singraph,color='g',label='sin(x)')
ax.plot(xvals,cosgraph,color='k',label='cos(x)')
ax.legend(loc='upper left')
ax.set_xbound(lower=-np.pi, upper=np.pi)

ax.plot()
```

```python
# Exercise 4

t,d,dy=np.loadtxt('day6ex4.txt',skiprows=5,unpack=True)

tsmooth = np.linspace(1,43,100)
baseline_y = (3+((1/2)*np.sin((np.pi*tsmooth)/5)))*tsmooth*np.exp(-tsmooth/10)

fig,ax = plt.subplots(1,1)
ax.plot(tsmooth, baseline_y, 'r-')
ax.errorbar(t, d, fmt='bo', yerr=dy, ecolor='black')
ax.set_ylabel('position (cm)')
ax.set_xlabel('time(s)')

ax.plot()
```

```python
# Exercise 6

d=np.array([0.38, 0.64, 0.91, 1.26, 1.41, 1.66, 1.90, 2.18])
f=np.array([1.4, 1.65, 3.0, 3.95, 4.3, 5.20, 6.85, 7.4])
df=np.array([ 0.4, 0.5, 0.4, 0.5, 0.6, 0.5, 0.5, 0.4])

coeff = np.polyfit(d,f,1)
poly = np.poly1d(coeff)
bestfit = poly(d)

fig,ax2 = plt.subplots(1,1)
ax2.errorbar(d, f, fmt='yD', yerr=df, ecolor='red',mec='red')
bestfitline = ax2.plot(d,bestfit,color='k',label="Slope = 3.509x - 0.3171")
ax2.legend(loc='upper left')
ax2.plot()

print('did not get the max/min (uncertainty of) slopes')
```
